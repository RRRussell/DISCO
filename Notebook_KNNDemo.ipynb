```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd

from torch.utils.data import Dataset, DataLoader

from utils import load_test_data, visualize_test_region, generate_training_samples
from dataset import STDataset
from baseline import RandomRegionBaseline, TissueSpecificRandomRegionBaseline, KNNClusteringBaseline
from evaluate import Evaluator
```

```python
all_test_items = load_test_data(num_holes=10)
```

```python
# Initialize dictionaries to store metrics
metrics = {
    'KNNBaseline': {
        'mse': [],
        'f1': [],
        'cosine_sim': [],
        'chamfer_dist': [],
        'emd': []
    }
}
```

```python
for i, test_item in enumerate(all_test_items):
    print(f"Test Area {i+1}:")
    print(f"  Dominant Tissue: {test_item.test_area.dominant_tissue}")
    print(f"  Number of cells in ground truth: {len(test_item.ground_truth.hole_cells)}")

    # Apply KNNBaseline
    KNN_baseline = KNNClusteringBaseline(test_item.adata, test_item.test_area)
    KNN_coords, KNN_gene_expressions = KNN_baseline.fill_region()

    # Evaluate predictions for KNNBaseline
    true_coords = test_item.ground_truth.hole_cells[['center_x', 'center_y']].values
    true_gene_expressions = test_item.ground_truth.gene_expression

    mse_knn, f1_knn, cosine_sim_knn = Evaluator.evaluate_expression(true_coords, true_gene_expressions, KNN_coords, KNN_gene_expressions)
    chamfer_dist_knn = Evaluator.chamfer_distance(true_coords, KNN_coords)
    emd_knn = Evaluator.calculate_emd(true_coords, KNN_coords)

    # Collect results for KNNBaseline
    metrics['KNNBaseline']['mse'].append(mse_knn)
    metrics['KNNBaseline']['f1'].append(f1_knn)
    metrics['KNNBaseline']['cosine_sim'].append(cosine_sim_knn)
    metrics['KNNBaseline']['chamfer_dist'].append(chamfer_dist_knn)
    metrics['KNNBaseline']['emd'].append(emd_knn)

   
```

```python
for method in metrics:
    print(f"Results for {method}:")
    for metric in metrics[method]:
        mean_value = np.mean(metrics[method][metric])
        std_value = np.std(metrics[method][metric])
        print(f"  {metric.capitalize()}: Mean = {mean_value:.4f}, Std = {std_value:.4f}")
```

```python
# Initialize a nested dictionary to store metrics for each tissue
tissues = ['striatum', 'cortical layer VI', 'cortical layer V', 'corpus callosum',
           'cortical layer II/III', 'olfactory region', 'pia mater', 'brain ventricle']
metrics = {
    tissue: {
        'KNNBaseline': {
            'mse': [], 
            'f1': [], 
            'cosine_sim': [], 
            'chamfer_dist': [], 
            'emd': []
        }
    } for tissue in tissues
}
```

```python
for i, test_item in enumerate(all_test_items):
    dominant_tissue = test_item.test_area.dominant_tissue
    print(f"Test Area {i+1}:")
    print(f"  Dominant Tissue: {dominant_tissue}")
    print(f"  Number of cells in ground truth: {len(test_item.ground_truth.hole_cells)}")
    
    # Apply KNNBaseline
    KNN_baseline = KNNClusteringBaseline(test_item.adata, test_item.test_area)
    KNN_coords, KNN_gene_expressions = KNN_baseline.fill_region()

    # Evaluate predictions for KNNBaseline
    true_coords = test_item.ground_truth.hole_cells[['center_x', 'center_y']].values
    true_gene_expressions = test_item.ground_truth.gene_expression

    mse_knn, f1_knn, cosine_sim_knn = Evaluator.evaluate_expression(true_coords, true_gene_expressions, KNN_coords, KNN_gene_expressions)
    chamfer_dist_knn = Evaluator.chamfer_distance(true_coords, KNN_coords)
    emd_knn = Evaluator.calculate_emd(true_coords, KNN_coords)

    # Collect results for KNNBaseline
    metrics[dominant_tissue]['KNNBaseline']['mse'].append(mse_knn)
    metrics[dominant_tissue]['KNNBaseline']['f1'].append(f1_knn)
    metrics[dominant_tissue]['KNNBaseline']['cosine_sim'].append(cosine_sim_knn)
    metrics[dominant_tissue]['KNNBaseline']['chamfer_dist'].append(chamfer_dist_knn)
    metrics[dominant_tissue]['KNNBaseline']['emd'].append(emd_knn)
```

```python
# Compute mean and std for each tissue
for tissue in metrics:
    print(f"Results for {tissue}:")
    for method in metrics[tissue]:
        print(f"  {method}:")
        for metric in metrics[tissue][method]:
            mean_value = np.mean(metrics[tissue][method][metric])
            std_value = np.std(metrics[tissue][method][metric])
            print(f"    {metric.capitalize()}: Mean = {mean_value:.4f}, Std = {std_value:.4f}")
```
